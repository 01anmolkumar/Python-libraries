```python
x=[1,2,3,4,5]
y1=[5,1,7,8,2]
y2 = [7,8,3,5,1]


plt.plot(x,y1,color='red',marker="o",linestyle='--',markersize=12,label='line y1')
plt.plot(x,y2,color='b',marker="o",linestyle='--',markersize=12,label='line y1')

plt.xlabel("x data axis")
plt.ylabel("y data axis")

plt.title('multiple lines in on plot')
plt.legend()

plt.show()



```

```python
#pichart>> to understand the percent sdist of data
langs=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=langs, autopct='%1.1f%%')
plt.title("pip chart")
plt.show()
```

```python
explode=(0.0,0.1,0.2)

langs=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=langs, autopct='%1.1f%%', explode=explode)
plt.title("pip chart")
plt.show()
```

```python
explode=(0.0,0.1,0.2)

langs=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=langs, autopct='%1.1f%%', explode=explode,shadow=True)
plt.title("pip chart")
plt.show()
```
