```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
#using graph object
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,3],mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,3],mode='lines'))
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip,mode='lines'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip,mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip,mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[3,8,5,2,6]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip,mode='markers',marker_size=12*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill, y=tips.tip,mode='markers',marker_size=10*tips['size'],z=tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1, 2, 3, 4, 5], y=[3, 8, 5, 2, 6], mode = 'lines', z = [1, 2, 7, 3, 5]))
```

```python
#using plotly express
x=[1,2,3,4]
y=[10,13,15,18]

fig=px.scatter(x,y)
fig.show()
```

```python
fig=px.line(x,y)
fig.show()
```

```python
x=['A','B','C']
y=[5,10,3]

fig=px.bar(x=x,y=y)
fig.show()
```

```python
x=[5,10,3,4,5,6,6,10,10,3,11,11,12]

fig=px.histogram(x)
fig.show()
```

```python
x=['A','B','C']
y=[5,10,3,25,12,33,100]

fig=px.box(x,y)
fig.show()
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
fig=px.imshow(data)
fig.show()
```

```python
#dubble chart 

x=[1,2,3,4]
y=[5,8,12,3]

fig = px.scatter(x,y, size=[30,10,50,200])
fig.show()
```

```python
#3d chart
x=[1,2,3,4]
y=[5,8,12,3]
z=[3,4,5,6]

fig = px.scatter_3d(x=x,y=y,z=z)
fig.show()
```

```python
#counter plot
x=[1,2,3,4,23,80]
y=[5,8,12,3,12,35,60]

fig =px.density_contour(x,y)
fig.show()
```
