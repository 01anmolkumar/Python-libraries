```python
#numerical data plotting
sns.relplot(x='total_bill',y='tip',data=df) # the relationship between two variables,by defoult it gives scatterplot,kind='line'
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='smoker',hue='time')
```

```python
#relplot to make a line plot>> we generate a data
time=np.arange(500)
value=np.random.randn(500).cumsum()
df1=pd.DataFrame({'time':time,'value':value})
```

```python
df1
```

```python
sns.relplot(x='time',y='value',kind='line',data=df1,sort=True)
```
